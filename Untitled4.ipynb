```python
class NoiseRemover:
    def __init__(self, data, window_size=30):
        self.data = data
        self.window_size = window_size

    def remove_noise(self):
        """
        Remove noise from the input time series using a moving average.

        Returns:
            pd.DataFrame: The DataFrame with noise removed.
        """
        smoothed_data = self.data.apply(lambda x: x.rolling(window=self.window_size).mean())
        return smoothed_data
data2 = NoiseRemover(data).remove_noise()
data2.columns = ["Open_s","High_s","Low_s","Close_s","Adj Close_s","Volume_s"]
data = pd.concat([data,data2],axis=1).dropna()
```

```python
import numpy as np
from backtesting import Backtest, Strategy
import pandas as pd

# Define the population size and number of generations
POP_SIZE = 100
NUM_GENERATIONS = 10

# Define the trading strategy
class TradingStrategy(Strategy):
    initsize = 0.01
    mysize = initsize
    last_trade_time = None
    atr = 22  # ATR calculation period
    prev_close = None
    prev_high = None
    prev_low = None
    n1 = 10
    n2 = 20
    trade_len = 50
    martingale_factor = 3  # Martingale factor
    mysize= 0.01
    def init(self):
        self.buy_threshold =np.zeros(len(self.data))
        self.sell_threshold = np.zeros(len(self.data))
        self.consecutive_losses = 0  # Track consecutive losses
        self.population = [
            (np.random.uniform(data.Close.min(), data.Close.max()), np.random.uniform(data.Close.min(), data.Close.max()))
            for _ in range(POP_SIZE)
        ]
        self.var = 0
    def calculate_atr(self, high, low, close):
        if self.prev_close is None or self.prev_high is None or self.prev_low is None:
            return high - low

        tr1 = abs(high - low)
        tr2 = abs(high - self.prev_close)
        tr3 = abs(low - self.prev_close)

        true_range = max(tr1, tr2, tr3)
        self.atr = ((self.atr * (self.atr_period - 1)) + true_range) / self.atr_period
        self.prev_close = close
        self.prev_high = high
        self.prev_low = low
        return self.atr

    
    def next(self):
        super().next()
        if len(self.trades) > 0:
            self.var = abs(np.sum([(trade.size*self.data.Close[-1])-trade.sl for trade in  self.trades])/self.equity)
        else:
            self.var = 0

        # Calculate ATR

#         # Check if enough time has passed since the last trade
#         if self.last_trade_time is not None:
#             time_since_last_trade = len(self.data) - self.last_trade_time
#         else:
#             time_since_last_trade = float('inf')  # Set to infinity initially
            
        # Run the genetic algorithm
        for generation in range(NUM_GENERATIONS):
            # Evaluate the fitness of each individual
            fitness_scores = [self.fitness(individual) for individual in self.population]

            tournament_size = 5
            parent_indices = np.random.choice(POP_SIZE, size=(POP_SIZE // 2, tournament_size), replace=True)
            parents = [self.population[i] for i in parent_indices.flatten() if i < len(self.population)]
            parents = sorted(parents, key=lambda x: self.fitness(x), reverse=True)[:POP_SIZE // 2]

            # Perform crossover
            offspring = []
            for parent1, parent2 in zip(parents[::2], parents[1::2]):
                crossover_point = np.random.randint(1, 2)
                child1 = (parent1[0], parent2[1]) if crossover_point == 0 else (parent2[0], parent1[1])
                child2 = (parent2[0], parent1[1]) if crossover_point == 0 else (parent1[0], parent2[1])
                offspring.append(child1)
                offspring.append(child2)

            # Mutate the offspring
            mutation_rate = 0.01
            for i in range(len(offspring)):
                if np.random.rand() < mutation_rate:
                    mutation_point = np.random.randint(2)
                    offspring[i] = (
                        np.random.uniform(self.data.Close.min(), self.data.Close.max()) if mutation_point == 0 else offspring[i][0],
                        np.random.uniform(self.data.Close.min(), self.data.Close.max()) if mutation_point == 1 else offspring[i][1]
                    )

            # Replace the population with the offspring
            self.population = offspring

        # Use the best individual from the final population
        best_buy_threshold, best_sell_threshold = max(self.population, key=self.fitness)
        self.buy_threshold = best_buy_threshold
        self.sell_threshold = best_sell_threshold
        atr = self.calculate_atr(self.data.High, self.data.Low, self.data.Close)
        self.mysize=self.calculate_position_size(self.var)
        self.vol20 = np.std(self.data.Close[-20:])/np.mean(self.data.Close[-20:])
        self.vol = np.std(self.data.Close)/np.mean(self.data.Close)
        
        if self.var < 1 and self.data.Close[-1] > self.buy_threshold  and self.vol20<0.02 and self.vol20 < self.vol:
            #self.buy(size=self.mysize, sl=self.data.Close[-1] - 3* atr, tp=self.data.Close[-1] + 5 * atr)
            self.buy(size=self.mysize, sl=self.data.Close[-1] *.9, tp=self.data.Close[-1] *1.1)
        if self.position.size > 0:
            if self.position.pl_pct < -0.15:
                self.position.close()
        if len(self.orders) > 0:
            for order in self.orders:
                if order.sl is not None:
                    if order.size*(abs(self.data.Close[-1] - order.sl)/self.data.Close[-1]) > 0.02:
                        order.cancel
        if self.var > 1:
            self.position.close(portion=0.1)
            
    def calculate_position_size(self,var):
        if abs(var)>0:
            size = min(0.05 *1/abs(var),.1)
        else:
            size = 0.05
        return size
        

    def fitness(self, individual):
        buy_threshold, sell_threshold = individual
        self.buy_threshold =buy_threshold
        self.sell_threshold = sell_threshold
        
        return self.sell_threshold/self.buy_threshold

bt = Backtest(data, TradingStrategy, cash=100000000000,margin=.5)
stats = bt.run()
print(stats)
```

```python
import numpy as np
import pandas as pd
from hmmlearn.hmm import GaussianHMM

# Load historical ETF price data
etf_prices = data

# Train HMM on each ETF return series
hmms = {}
for etf in etf_prices.columns:
    returns = etf_prices[etf].pct_change().dropna()
    hmm = GaussianHMM(n_components=3, covariance_type="diag", n_iter=1000)
    hmm.fit(returns.values.reshape(-1, 1))
    hmms[etf] = hmm

# Infer hidden states for each ETF
etf_states = etf_prices.copy()
for etf in etf_prices.columns:
    returns = etf_prices[etf].pct_change().dropna().values.reshape(-1, 1)
    _, etf_states[etf] = hmms[etf].decode(returns)

# Compute state-dependent weights
state_weights = etf_states.groupby(etf_states, axis=1).mean()

# Construct portfolio
portfolio = etf_prices.iloc[-1] * 0
for etf in etf_prices.columns:
    portfolio += state_weights[etf] * etf_prices[etf].iloc[-1]

print(portfolio)
```
