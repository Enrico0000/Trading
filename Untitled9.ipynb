```python
#!pip install backtesting
#!pip install yfinance
from backtesting import Strategy
from backtesting import Backtest
from backtesting.test import SMA
from backtesting.lib import crossover
import pandas as pd
import numpy as np
import yfinance as yf
import warnings
class Stationarity:
    def __init__(self,data):
        self.data = np.log(data)
        self.col = [x for x in self.data]
        self.index = [y for y in self.data.T]
        self.close = self.data.Close
        self.Open_diff = self.data.Open - self.data.Close
        self.High_diff = self.data.High - self.data.Close
        self.Low_diff = self.data.Low - self.data.Close
        self.data_st = None
    def _stationarity(self):
        drift = (self.close[-1]-self.close[0])/(len(self.close)-1)
        d = np.array([(self.close[i]-drift*i) for i in range(len(self.close))])
        return pd.Series(d.reshape(-1))
    def execute(self):
        Close = self._stationarity()
        Open = pd.Series(np.array(self.Open_diff) + np.array(Close))
        High = pd.Series(np.array(self.High_diff) + np.array(Close))
        Low = pd.Series(np.array(self.Low_diff) + np.array(Close))
        self.data_st = pd.concat([Open,High,Low,Close],axis=1)
        self.data_st.columns = columns = self.col
        self.data_st.index = self.index
        return np.exp(self.data_st)
warnings.filterwarnings("ignore")
```

```python
def kf(measurements):
    measurements = np.array(measurements)
    # Define system matrices
    F = np.array([[1, 1], [0, 1]])  # State transition matrix
    H = np.array([[1, 0]])           # Observation matrix
    Q = np.eye(2) * 0.01             # Process noise covariance matrix
    R = np.array([[0.1]])            # Measurement noise covariance matrix
    # Initialize state and covariance
    x_hat = np.zeros((2, 1))         # Initial state estimate
    P = np.eye(2) * 0.1              # Initial error covariance


#     # Generate true state and measurements
#     true_state = [np.random.normal(0, 0.1) for _ in range(100)]  # True state trajectory
#     measurements = [state + np.random.normal(0, 0.1) for state in true_state]  # Add noise to true state for measurements


    # Initialize lists to store estimated state and covariance
    estimated_state = []


    # Iterate over measurements
    for z in measurements:
        # Prediction step
        x_hat = F @ x_hat
        P = F @ P @ F.T + Q

        # Update step
        K = P @ H.T @ np.linalg.inv(H @ P @ H.T + R)
        y = z - H @ x_hat
        x_hat = x_hat + K @ y
        P = (np.eye(2) - K @ H) @ P

        # Store estimated state
        estimated_state.append(x_hat[0, 0])
    return pd.Series(np.array(estimated_state).reshape(-1))
```
