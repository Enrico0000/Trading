```python
import numpy as np
import matplotlib.pyplot as plt

def simulate_brownian_motion(T, dt, sigma):
    """
    Simulate a Brownian motion process.
    
    Parameters:
    T (float): Total time of the simulation.
    dt (float): Time step.
    sigma (float): Volatility of the process.
    
    Returns:
    np.array: Array of Brownian motion values.
    """
    N = int(T / dt)
    t = np.linspace(0, T, N)
    dW = sigma * np.random.normal(0, np.sqrt(dt), N)
    W = np.cumsum(dW)
    return W

# Example usage
T = 100.0  # Total time
dt = 0.01  # Time step
sigma = 1.0  # Volatility

W = simulate_brownian_motion(T, dt, sigma)
```

```python
import pandas as pd
import yfinance as yf
class NoiseRemover:
    def __init__(self, data, window_size=5):
        self.data = data
        self.window_size = window_size

    def remove_noise(self):
        """
        Remove noise from the input time series using a moving average.

        Returns:
            pd.DataFrame: The DataFrame with noise removed.
        """
        smoothed_data = self.data.apply(lambda x: x.rolling(window=self.window_size).mean())
        return smoothed_data
```

```python
import numpy as np
from itertools import product

class TimeVaryingParameters:
    def __init__(self, mu0, mu1, sigma_mu, kappa0, kappa1, sigma_kappa, sigma0, sigma1, sigma_sigma, lamb0, lamb1, sigma_lambda, delta0, delta1, sigma_delta):
        self.mu0 = mu0
        self.mu1 = mu1
        self.sigma_mu = sigma_mu
        self.kappa0 = max(kappa0, 0.01)
        self.kappa1 = max(kappa1, 0.01)
        self.sigma_kappa = sigma_kappa
        self.sigma0 = sigma0
        self.sigma1 = sigma1
        self.sigma_sigma = sigma_sigma
        self.lamb0 = lamb0
        self.lamb1 = lamb1
        self.sigma_lambda = sigma_lambda
        self.delta0 = delta0
        self.delta1 = delta1
        self.sigma_delta = sigma_delta

    def get_parameters(self, t):
        mu = np.clip(self.mu0 + self.mu1 * t + self.sigma_mu * np.random.normal(), 0, 1)
        kappa = self.kappa0 + self.kappa1 * t + self.sigma_kappa * np.random.normal()
        sigma = self.sigma0 + self.sigma1 * t + self.sigma_sigma * np.random.normal()
        lamb = max(0.01, self.lamb0 + self.lamb1 * t + self.sigma_lambda * np.random.normal())  # Ensure lamb is positive
        delta = self.delta0 + self.delta1 * t + self.sigma_delta * np.random.normal()
        return mu, kappa, sigma, lamb, delta

class StochasticVolatility:
    def __init__(self, kappa_v, theta_v, sigma_v, V0):
        self.kappa_v = kappa_v
        self.theta_v = theta_v
        self.sigma_v = max(sigma_v, 0.01)
        self.V0 = V0

    def update_volatility(self, V, dt):
        dW_v = np.random.normal(scale=np.sqrt(dt))
        dV = self.kappa_v * (self.theta_v - V) * dt + self.sigma_v * np.sqrt(max(V, 0.01)) * dW_v
        return V + dV

class RegimeSwitch:
    def __init__(self, num_regimes, transition_probs):
        self.num_regimes = num_regimes
        self.transition_probs = transition_probs
        self.current_regime = 0

    def switch_regime(self):
        self.current_regime = np.random.choice(self.num_regimes, p=self.transition_probs[self.current_regime])

def generate_time_series(time_varying_params, stochastic_vol, regime_switch, initial_value, T, dt, N):
    O = np.zeros(N + 1)
    H = np.zeros(N + 1)
    L = np.zeros(N + 1)
    C = np.zeros(N + 1)

    O[0] = initial_value
    H[0] = initial_value
    L[0] = initial_value
    C[0] = initial_value

    V = stochastic_vol.V0

    for t in range(1, N + 1):
        regime_switch.switch_regime()
        mu, kappa, sigma, lamb, delta = time_varying_params.get_parameters(t * dt)
        dW = np.sqrt(dt) * np.random.normal()
        dN = np.random.poisson(lamb * dt)
        V = stochastic_vol.update_volatility(V, dt)

        # Simulate the OHLC prices
        O[t] = np.clip(O[t - 1] + kappa * (mu - O[t - 1]) * dt + sigma * O[t - 1] * dW + (np.exp(delta * np.random.normal()) - 1) * dN, 0, 1)
        H[t] = max(H[t - 1], O[t], C[t - 1] * (1 + sigma * dW + (np.exp(delta * np.random.normal()) - 1) * dN))
        L[t] = min(L[t - 1], O[t], C[t - 1] * (1 + sigma * dW + (np.exp(delta * np.random.normal()) - 1) * dN))
        C[t] = C[t - 1] * (1 + sigma * dW + (np.exp(delta * np.random.normal()) - 1) * dN)

    return C# O, H, L, 

def objective_function(params, open_prices, high_prices, low_prices, close_prices):
    time_varying_params = TimeVaryingParameters(*params[:15])
    stochastic_vol = StochasticVolatility(*params[15:19])
    regime_switch = RegimeSwitch(num_regimes=3, transition_probs=[[params[19], params[20], 1 - params[19] - params[20]],
                                                                  [params[21], params[22], 1 - params[21] - params[22]],
                                                                  [params[23], params[24], 1 - params[23] - params[24]]])

    # Check if the transition probabilities are valid
    if any(p < 0 or p > 1 for p in params[19:25]):
        return np.inf  # Return a very large value to penalize invalid parameter values

    try:
        C = generate_time_series(time_varying_params, stochastic_vol, regime_switch, open_prices[0], T=1, dt=0.01, N=100)
    except ValueError:
        return np.inf  # Return a very large value to penalize invalid parameter values

#     mse_o = np.mean((O - open_prices[:len(O)]) ** 2)
#     mse_h = np.mean((H - high_prices[:len(H)]) ** 2)
#     mse_l = np.mean((L - low_prices[:len(L)]) ** 2)
    mse_c = np.mean((C - close_prices[:len(C)]) ** 2)

    return  mse_c #+mse_o + mse_h + mse_l

from statsmodels.tsa.arima.model import ARIMA
import numpy as np
from scipy.stats import boxcox
from sklearn.preprocessing import MinMaxScaler
data = yf.download("SPY").drop(["Adj Close","Volume"],axis=1)
data = np.log(data)
differenze = pd.concat([data["Open"] - data["Close"],data["High"] - data["Close"],data["Low"]-data["Close"]],axis=1)
differenze.columns = ["Open","High","Low"]
data["Close"] = (data["Close"].diff().fillna(0) + data["Close"][0])
data["High"] = differenze["High"] + data["Close"]
data["Low"] = differenze["Low"] + data["Close"]
data["Open"] = differenze["Open"] + data["Close"]
differenze
scaler = MinMaxScaler()

# Create an instance of the NoiseRemover class
#data = np.log(data[["Open", "High", "Low", "Close"]])
#data = NoiseRemover(data[["Open", "High", "Low", "Close"]]).remove_noise()
# model1 = ARIMA(data["Open"], order=(0,1,0))
# model1 = model1.fit()
# data1 = model1.resid[1:]+100
# model2 = ARIMA(data["High"], order=(0,1,0))
# model2 = model2.fit()
# data2 = model2.resid[1:]+100
# model3 = ARIMA(data["Low"], order=(0,1,0))
# model3 = model3.fit()
# data3 = model3.resid[1:]+100
# model4 = ARIMA(data["Close"], order=(0,1,0))
# model4 = model4.fit()
# data4 = model4.resid[1:]+100
 
# open_prices = data1.values
# high_prices = data2.values
# low_prices = data3.values
# close_prices = data4.values

# open_prices = scaler.fit_transform(data["Open"].values.reshape(-1,1)).reshape(-1)
# high_prices = scaler.fit_transform(data["High"].values.reshape(-1,1)).reshape(-1)
# low_prices = scaler.fit_transform(data["Low"].values.reshape(-1,1)).reshape(-1)
# close_prices = scaler.fit_transform(data["Close"].values.reshape(-1,1)).reshape(-1)
open_prices = data["Open"].values
high_prices = data["High"].values
low_prices = data["Low"].values
close_prices = data["Close"].values

# Genetic algorithm optimization
population_size = 100
num_generations = 100

# Initialize the population randomly
population = np.random.uniform(low=0, high=1, size=(population_size, 25))

for generation in range(num_generations):
    # Evaluate the fitness of each individual in the population
    fitness = [objective_function(individual, open_prices, high_prices, low_prices, close_prices) for individual in population]

    # Select the parents for crossover
    parents = np.argsort(fitness)[:population_size // 2]

    # Perform crossover to generate offspring
    offspring = []
    for i in range(0, len(parents), 2):
        if i + 1 < len(parents):
            parent1 = population[parents[i]]
            parent2 = population[parents[i + 1]]
            crossover_point = np.random.randint(1, 25)
            child1 = np.concatenate((parent1[:crossover_point], parent2[crossover_point:]))
            child2 = np.concatenate((parent2[:crossover_point], parent1[crossover_point:]))
            offspring.append(child1)
            offspring.append(child2)

    # Mutate the offspring
    mutation_rate = 0.1
    offspring = np.array(offspring)
    for i, j in product(range(len(offspring)), range(len(offspring[0]))):
        if np.random.rand() < mutation_rate:
            offspring[i][j] = np.random.uniform(low=0, high=1)
    print(generation)
    # Replace the population with the offspring
    population = offspring

# Print the best individual
best_individual = population[np.argmin([objective_function(individual, open_prices, high_prices, low_prices, close_prices) for individual in population])]
print("Best individual:", best_individual)

# Generate the time series using the best individual
time_varying_params = TimeVaryingParameters(*best_individual[:15])
stochastic_vol = StochasticVolatility(*best_individual[15:19])
regime_switch = RegimeSwitch(num_regimes=3, transition_probs=[[best_individual[19], best_individual[20], 1 - best_individual[19] - best_individual[20]],
                                                              [best_individual[21], best_individual[22], 1 - best_individual[21] - best_individual[22]],
                                                              [best_individual[23], best_individual[24], 1 - best_individual[23] - best_individual[24]]])
```
