```python
def kf(measurements):
    measurements = np.array(measurements)
    # Define system matrices
    F = np.array([[1, 1], [0, 1]])  # State transition matrix
    H = np.array([[1, 0]])           # Observation matrix
    Q = np.eye(2) * 0.01             # Process noise covariance matrix
    R = np.array([[0.1]])            # Measurement noise covariance matrix
    # Initialize state and covariance
    x_hat = np.zeros((2, 1))         # Initial state estimate
    P = np.eye(2) * 0.1              # Initial error covariance


#     # Generate true state and measurements
#     true_state = [np.random.normal(0, 0.1) for _ in range(100)]  # True state trajectory
#     measurements = [state + np.random.normal(0, 0.1) for state in true_state]  # Add noise to true state for measurements


    # Initialize lists to store estimated state and covariance
    estimated_state = []


    # Iterate over measurements
    for z in measurements:
        # Prediction step
        x_hat = F @ x_hat
        P = F @ P @ F.T + Q

        # Update step
        K = P @ H.T @ np.linalg.inv(H @ P @ H.T + R)
        y = z - H @ x_hat
        x_hat = x_hat + K @ y
        P = (np.eye(2) - K @ H) @ P

        # Store estimated state
        estimated_state.append(x_hat[0, 0])
    return pd.Series(np.array(estimated_state).reshape(-1))
```

```python
from scipy.stats import shapiro
from scipy.stats import norm
def positive_normal():
    num = 4
    while num > 3.5:
        num = np.random.normal()
        if num > 0:
            return num
        if num < 0:
            return -num
def Logit(regr_sample):
    regr_sample = np.log(regr_sample) - np.log(1-regr_sample)
    return regr_sample
def normalization(regr_sample):
    std = regr_sample.std()
    regr_sample = regr_sample/regr_sample.std()
    while regr_sample.max() >= 3.5 or regr_sample.min() <= -3.5:
        regr_sample[(regr_sample > 3.5)] = positive_normal() 
        regr_sample[(regr_sample < -3.5)] = -positive_normal()
    frac_rank = (np.argsort(np.argsort(regr_sample)) + 0.5) / (len(regr_sample) + 0.5)
    return  pd.Series(norm.ppf(frac_rank)) * std
```
